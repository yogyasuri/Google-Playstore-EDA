> #### 3.2.2 Clean the `Size` column.

```python
# check the unique values
df['Size'].unique()
```

```python
# By checking value counts will help to explore the more unique values
df['Size'].value_counts().head()
```

### Solution:
To prepare the `Size` column for analysis, we must:
1. Convert the unit of measurement (M,G or K) to a consistent format, such as bytes.
2. Remove the unit suffix ("M","G" or "K") from the values.
3. Change the data type of the values to numeric.

```python
df['Size'].isnull().sum()
```

```python
# Remove commas from values
df['Size'] = df['Size'].str.replace(',', '')
```

> Let's Convert "M","G" and "k" units in storage values to bytes by removing the respective letters and converting the remaining numeric data to a numeric data type.

```python
# Change the "size" column to a numeric format by performing the following transformations:
#  1. If the value contains the letter "k", multiply it by 1024.
#  2. If the value contains the letter "M", multiply it by 1024*1024.
#  2. If the value contains the letter "G", multiply it by 1024*1024*1024.
#  3.this function will convert the size column to numeric
def convert_size(size):
    # add function details here
    '''
    This function transforms the "size" column into a numeric format.
    It multiplies the values by 1024 if there is a "k" in the value, and by 1024*1024 if there is an "M" in the value.
    '''
def convert_size(size):
    if isinstance(size, str):
        size = size.replace(',', '')  # Remove commas from the value
        if 'k' in size:
            return float(size.replace('k', '')) * 1024
        elif 'M' in size:
            return float(size.replace('M', '')) * 1024 * 1024
        elif 'G' in size:
            return float(size.replace('G', '')) * 1024 * 1024 * 1024
        elif 'Varies with device' in size:
            return np.nan
    return size
# Apply the function to the Size
df['Size'] = df['Size'].apply(convert_size)
```

```python
# Let's rename the Size column to Size_in_bytes
df.rename(columns={'Size': 'Size_in_bytes'}, inplace=True)
```

```python
# making a new column called 'Size in Mb' which will have the size in MB
df['Size_in_Mb'] = df['Size_in_bytes'].apply(lambda x: x/(1024*1024))
```

- Each value has been converted to its byte equivalent, with the `"M"` ,`"G"` and `"K"` suffixes removed.
- The resulting values are all numeric data types.
- The `"Varies with device"` value was originally a string.
- For data consistency, we have temporarily replaced it with a null value. We can assign appropriate values later as needed.

### Obsrvations:
1. The `Rating Count` column cannot be imputed directly because it is closely related to the `Installs` column.

    - Imputing missing values with the mean or median could result in misleading information because some apps may have zero `Installs`. In such cases, the imputed `Rating Count` would replace the actual zero value, potentially underestimating the true install count.

    - To determine this relationship, we will perform a `"Pearson correlation test"` to statistically evaluate the strength and direction of the correlation.

2. To ensure data accuracy, we need to eliminate rows with missing entries, such as those for `Released`, `Rating`,`Minimum Android`, `Currency`, `Minimum Installs`, `Installs`, `Installs_category`, and `App Name`.

3. We've already handled the `Varies with device` values in the Size columns by converting them to null. We don't need to impute these values right now because each app has a unique size that's hard to predict accurately.

```python
# length of dataset before removing null values
print(f"Length of the dataframe before removing null values: {len(df)}")

'''
remove the rows having null values in  `Released`, `Rating`, `Minimum Android`, `Currency`,
 `Minimum Installs`, `Installs`, `Installs_category`, and `App Name`.
'''
df.dropna(subset=['Released', 'Minimum Android', 'Currency','Rating',
                  'Minimum Installs', 'Installs', 'Installs_category', 'App Name'], inplace=True)

# length of dataset after removing null values
print(f"Length of the dataframe after removing null values: {len(df)}")
```
